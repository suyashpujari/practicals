```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    # e_x = np.exp(x - np.max(x))
    # return e_x / e_x.sum()
    return np.exp(x) / np.sum(np.exp(x), axis=0)

# Input range
x = np.linspace(-5, 5, 100)

# Compute activation function values
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_softmax = softmax(x)

# Plotting
plt.figure(figsize=(10, 6))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, 'b-', label='Sigmoid')
plt.title('Sigmoid Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# ReLU
plt.subplot(2, 2, 2)
plt.plot(x, y_relu, 'r-', label='ReLU')
plt.title('ReLU Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Tanh
plt.subplot(2, 2, 3)
plt.plot(x, y_tanh, 'g-', label='Tanh')
plt.title('Tanh Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Softmax
plt.subplot(2, 2, 4)
plt.plot(x, y_softmax, 'm-', label='Softmax')
plt.title('Softmax Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
